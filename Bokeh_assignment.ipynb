```python
# Q.1 Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

from bokeh.plotting import figure, show
import numpy as np

# Generate data for the sine wave
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = np.sin(x)  # y-values as the sine of x

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', plot_height=400, plot_width=800)

# Add a line renderer for the sine wave
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the plot
show(p)
```

```python
# Q.2 Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np
import pandas as pd

# Generate random data
np.random.seed(42)
n = 100
x = np.random.rand(n) * 10  # Random x values
y = np.random.rand(n) * 10  # Random y values
sizes = np.random.randint(10, 50, size=n)  # Random sizes for markers
colors = np.random.choice(['#FF6347', '#4682B4', '#32CD32', '#FFD700'], size=n)  # Random colors

# Create a DataFrame to store the data
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create ColumnDataSource from the DataFrame
source = ColumnDataSource(data)

# Create the scatter plot
plot = figure(title="Scatter Plot with Random Sizes and Colors",
              x_axis_label='X', y_axis_label='Y', 
              width=800, height=600)

# Add scatter plot
plot.scatter(x='x', y='y', size='sizes', color='colors', source=source, legend_field="colors", fill_alpha=0.6)

# Show the plot
show(plot)
```

```python
# Q.3  Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]


from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource
import pandas as pd

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a DataFrame
df = pd.DataFrame({'Fruit': fruits, 'Count': counts})

# Create ColumnDataSource from DataFrame
source = ColumnDataSource(df)

# Create a figure for the bar chart
plot = figure(x_range=fruits, title="Fruit Count Bar Chart",
              x_axis_label='Fruit', y_axis_label='Count',
              width=800, height=600)

# Add bars to the plot
plot.vbar(x='Fruit', top='Count', width=0.5, source=source, legend_field="Fruit", color="green", alpha=0.7)

# Show the plot
output_file("fruit_count_bar_chart.html")  # Save the output as an HTML file
show(plot)
```

```python
# Q.4  . Create a Bokeh histogram to visualize the distribution of the given data.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file

# Generate random data
data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure for the histogram
plot = figure(title="Histogram of Random Normal Data", 
              x_axis_label='Value', y_axis_label='Frequency', 
              tools="pan, box_zoom, reset", width=800, height=600)

# Add the bars to the plot
plot.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
          fill_color="skyblue", line_color="white", alpha=0.7)

# Show the plot
output_file("normal_distribution_histogram.html")  # Save as an HTML file
show(plot)
```

```python
# Q.5   Create a Bokeh heatmap using the provided dataset.


import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.layouts import column

# Provided dataset
data_heatmap = np.random.rand(10, 10)  # 10x10 random data
x = np.linspace(0, 1, 10)  # x coordinates (10 points between 0 and 1)
y = np.linspace(0, 1, 10)  # y coordinates (10 points between 0 and 1)
xx, yy = np.meshgrid(x, y)  # Generate a grid from x and y

# Flatten the arrays for plotting
x = xx.flatten()
y = yy.flatten()
values = data_heatmap.flatten()

# Create a figure for the heatmap
p = figure(width=600, height=600, title="Heatmap Example",
           x_axis_label='X', y_axis_label='Y', tools="hover,pan,box_zoom,reset")

# Use a linear colormap to map data values to colors
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(values), high=max(values))

# Plot the squares of the heatmap using quad glyphs
p.quad(x=x, y=y, width=0.1, height=0.1, color=mapper, source={'value': values})

# Add color bar for the heatmap
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
output_file("heatmap_example.html")  # Save the output as an HTML file
show(p)
```
