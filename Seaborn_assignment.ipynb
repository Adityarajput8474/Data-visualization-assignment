```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Q.1 Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.



np.random.seed(0)
x = np.random.rand(100) * 10  # Random data for variable x
y = 2.5 * x + np.random.randn(100) * 2  # y is linearly related to x with some noise

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='b', alpha=0.7, label='Data points')
plt.title('Scatter Plot of Synthetic Data', fontsize=14)
plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Q.2 Generate a dataset of random numbers. visualize the distribution of a numerical variable.

# Generate a dataset of random numbers (e.g., 1000 random numbers from a normal distribution)
np.random.seed(0)
data = np.random.randn(1000)  # Normally distributed data with mean 0 and std dev 1

# Create the histogram to visualize the distribution
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of a Numerical Variable', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Q.3  Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.


# Create a dataset of categories and corresponding numerical values
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=5)  # Random values between 10 and 100 for each category

# Create the bar plot
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='lightcoral', edgecolor='black')
plt.title('Comparison of Categories Based on Numerical Values', fontsize=14)
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Q.4 . Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

# Create the dataset
categories = ['Category A', 'Category B', 'Category C', 'Category D']
num_samples = 100

data = {
    'Category': np.random.choice(categories, size=num_samples),
    'Value': np.random.randn(num_samples) * 10 + 50  # Random numerical values with some variation
}

df = pd.DataFrame(data)

# Create a boxplot to visualize the distribution
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Distribution of Numerical Values Across Categories')
plt.show()
```
