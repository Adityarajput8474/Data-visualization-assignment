```python
# Q.1 . Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set the random seed for reproducibility
np.random.seed(30)

# Generate data
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Extract the X, Y, Z values
x = df['X']
y = df['Y']
z = df['Z']

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot the data points
ax.scatter(x, y, z, c='b', marker='o', alpha=0.6)

# Set labels for each axis
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Set title for the plot
ax.set_title('3D Scatter Plot of Randomly Generated Data')

# Show the plot
plt.show()
```

```python
# Q.2   Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories.


import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(15)

# Creating the student grades dataset
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}

df = pd.DataFrame(data)

# Create the violin plot
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.violinplot(x='Grade', y='Score', data=df, inner='quart', palette='muted')

# Set title and labels
plt.title('Distribution of Scores across Grade Categories')
plt.xlabel('Grade Category')
plt.ylabel('Score')

# Show the plot
plt.show()
```

```python
# Q.3 . Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(20)

# Creating the sales dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Pivot the data for heatmap (Sales data per month and day)
heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.1f', cbar_kws={'label': 'Sales Amount'}, linewidths=0.5)

# Set title and labels
plt.title('Sales Heatmap Across Months and Days')
plt.xlabel('Month')
plt.ylabel('Day of the Month')

# Show the plot
plt.show()
```

```python
# Q.4  Using the given x and y data, generate a 3D surface plot to visualize the function 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create meshgrid for x and y
x, y = np.meshgrid(x, y)

# Compute z values based on the function
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the arrays to create the dataframe
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

# Create a pandas DataFrame
df = pd.DataFrame(data)

# Plot the 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create the surface plot
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Set labels for the axes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Set title
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

# Add color bar
fig.colorbar(surf)

# Show the plot
plt.show()
```

```python
# Q.5  . Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(25)

# Creating the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plotting the bubble chart
plt.figure(figsize=(10, 6))

# Use 'Population' for the y-axis, 'GDP' for the x-axis, and 'Population' for the size of the bubbles
plt.scatter(df['GDP'], df['Population'], s=df['Population'] * 10, alpha=0.6, c=df['Population'], cmap='viridis')

# Adding labels and title
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')
plt.title('Bubble Chart: GDP vs Population by Country')

# Annotate each point with the country's name
for i, country in enumerate(df['Country']):
    plt.text(df['GDP'][i] + 20, df['Population'][i], country, fontsize=9)

# Show the plot
plt.colorbar(label='Population')
plt.grid(True)
plt.show()

```
